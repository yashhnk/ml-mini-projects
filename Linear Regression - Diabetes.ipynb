```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
x = diabetes.data[:, 2]
y = diabetes.target
```

```python
plt.scatter(x, y, color="blue")
plt.xlabel("BMI")
plt.ylabel("diabetes progression")
plt.title("bmi vs diabetes progression")
plt.grid(True)
plt.show()
```

```python
m=0
b=0
epochs=2000
learning_rate=0.01
def compute_cost(x,y,m,b):
    n=len(x)
    total_error=0
    for i in range(n):
        prediction = m*x[i] + b
        error = prediction - y[i]
        total_error += error**2
    return total_error/n
```

```python
def gradient_descent(x,y,m,b,epochs,learning_rate):
    n = len(x)
    for j in range(epochs):
        dm = 0
        db = 0
        for i in range(n):
            prediction = m*x[i]+b
            error = prediction - y[i]
            dm += error*x[i]
            db += error
        dm = (2/n)*dm
        db = (2/n)*db
        m = m - learning_rate*dm
        b = b- learning_rate*db
    return m,b
```

```python
m,b = gradient_descent(x,y,m,b,epochs,learning_rate)
print(f"Final m :{m}, Final(b) : {b}")
```

```python
y_prediction =x*m + b
plt.scatter(x,y,color="blue", label="data")
plt.plot(x,y_prediction, color="red", label="best fit")
plt.xlabel("BMI")
plt.ylabel("Diabetes Progression")
plt.title("Linear Regression on Health Dataset")
plt.grid(True)
plt.legend()
plt.show()
```
